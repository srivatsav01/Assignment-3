```python

import numpy as np
from scipy.stats import bernoulli

set_size=100000

th_pr_A=1/2
th_pr_B=1/8



#simulations by bernoulli r.v.
B_set= bernoulli.rvs(size= sample_size, p = th_pr_B)
A_set= bernoulli.rvs(size= sample_size, p = th_pr_A)

#calculating the cases we want 
B_ds = np.nonzero(B_set == 1)
A_ds = np.nonzero(A_set == 1)


num_A= np.size(A_ds)
num_B= np.size(B_ds)

#calculating probability via simulations
sim_pr_A=num_A/set_size
sim_pr_B=num_B/set_size
sim_pr_AgivenC=sim_pr_A/(sim_pr_A+sim_pr_B)

th_AgivenC=th_pr_A/(th_pr_A+th_pr_B)



print("P(AB)simulated Value  = ", sim_pr_AgivenC, " while theoretical value = ",th_AgivenC,)
```
